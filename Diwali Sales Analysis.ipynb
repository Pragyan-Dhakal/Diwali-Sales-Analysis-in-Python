```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
# Replaceing 1 with 'Married' and 0 with 'Unmarried' in the Marital_Status column
df['Marital_Status'] = df['Marital_Status'].replace({1: 'Married', 0: 'Unmarried'})


df.info()
```

```python
df.isnull().sum()
```

```python
#Deleting a null value
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender', data=df, palette='husl', width=0.5)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count of Gender')
plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

ax = sns.barplot(x='Gender', y='Amount', data=sales_gen, palette='Set1', width=0.5)  # Try 'husl', 'pastel', etc.

plt.title ('Sales by Gender',fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.show()
```

## From above graphs we can see that the most of the buyers are females and even the perchasing power of females are greater than men.

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender', palette='Set1') 


for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Count of Age Groups', fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='Set1')  # Try 'husl', 'pastel', etc.
plt.xticks(rotation=45)
plt.title('Sales by Age Group', fontdict={'fontsize': 15, 'fontweight': 'bold'})

plt.show()
```

## From above graphs we can see that most of the buyers are of age group between 26-35 age

# State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

sns.set(rc={'figure.figsize':(15, 8)})
sns.barplot(x='State', y='Amount', data=sales_state, palette='Set1')  # Try 'husl', 'pastel', etc.
plt.xticks(rotation=45)

plt.title('Sales by State', fontdict={'fontsize': 25, 'fontweight': 'bold'})
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values('Orders', ascending=False)

sns.set(rc={'figure.figsize':(15, 8)})
sns.barplot(x='State', y='Orders', data=sales_state, palette='Set1')  # Try 'husl', 'pastel', etc.
plt.xticks(rotation=45)

plt.title('Orders by State', fontdict={'fontsize': 25, 'fontweight': 'bold'})
plt.show()
```

## From the above graph we can see that most of orders and total sales/Amount are from Utter Pradesh, Maharashtra and Karnataka

## Marital Status

```python
ax = sns.countplot(x='Marital_Status', data=df, palette='husl', width=0.5)  

plt.title('Marital Status Count', fontdict={'fontsize': 15, 'fontweight': 'bold'})

sns.set(rc={'figure.figsize':(7, 5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_ms = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

sns.set(rc={'figure.figsize':(6, 5)})
sns.barplot(x='Marital_Status', y='Amount', hue = 'Gender', data=sales_ms, palette='Set1')  # Try 'husl', 'pastel', etc.


plt.title('Sales by Marital Status and Gender ', fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.show()
```

# From above graph we can see that most of the buyers are married women and they have high purchasing power.

# Occupation

```python
ax = sns.countplot(x='Occupation', data=df, palette='husl', width=0.5)  

plt.title('Occupation Count', fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.xticks(rotation=45)

sns.set(rc={'figure.figsize':(15, 5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_ms = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

sns.set(rc={'figure.figsize':(15, 6)})
sns.barplot(x='Occupation', y='Amount', data=sales_ms, palette='Set1')  # Try 'husl', 'pastel', etc.
plt.xticks(rotation=45)


plt.title('Sales by Occupation', fontdict={'fontsize': 25, 'fontweight': 'bold'})
plt.show()
```

## From above graphs we can see that most of the buyers are workers in IT, Healthcare Sector and Aviation.

# Product Category

```python
# Get the top 5 product categories by count
top_categories = df['Product_Category'].value_counts().head(10).index
filtered_df = df[df['Product_Category'].isin(top_categories)]

sns.set(rc={'figure.figsize': (10, 5)})  # Set figure size

# Plot the countplot for the top 5 product categories
ax = sns.countplot(x='Product_Category', data=filtered_df, palette='husl', width=0.5)

plt.title('Top 10 Product Categories Count', fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

for bars in ax.containers:
    ax.bar_label(bars)  # Add bar labels

plt.show()
```

```python
# Get the top 10 product categories by count
top_categories = df['Product_Category'].value_counts().head(10).index
filtered_df = df[df['Product_Category'].isin(top_categories)]

# Group by Product_Category and calculate the sum of Amount for the top categories
sales_ms = filtered_df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

sns.set(rc={'figure.figsize': (15, 6)})

# Plot the barplot for the top 10 product categories
sns.barplot(x='Product_Category', y='Amount', data=sales_ms, palette='Set1')  # Try 'husl', 'pastel', etc.
plt.xticks(rotation=45)

plt.title('Sales by Top 10 Product Categories', fontdict={'fontsize': 25, 'fontweight': 'bold'})
plt.show()
```

## From above graphs can see that most of the sold products are food, Clothing and Electronics category.

## Conclusion:

In our Diwali Sales Analysis, we found that approximately 60% of transactions were made by females, while 40% came from males. Interestingly, even with fewer transactions, males generated about 35% more revenue, indicating a higher average sales value per transaction. These insights, visualized through customized count and bar plots, underscore the importance of targeting marketing strategies to cater to differing purchasing behaviors across genders.
